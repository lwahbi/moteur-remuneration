```python
import os
import shutil
import random
from google.colab import drive

drive.mount('/content/drive')

# Définition des paramètres initiaux pour la génération des données SQL
nombre_enregistrements = 2000000  # Nombre d'enregistrements à générer
nombre_codes_es = 10000  # Nombre de codes ES différents
nombre_code_oper_max = 1000  # Nombre maximum de code_oper différents

codes_es = [f'ES{i}' for i in range(1, nombre_codes_es + 1)]  # Codes ES possibles pour la variation

# Générer une liste aléatoire de code_oper
codes_oper = [f'CodeOper_{i}' for i in range(1, nombre_code_oper_max + 1)]

# Chemin de destination sur Google Drive
chemin_google_drive = '/content/drive/My Drive/clients_transactions_insert.sql'

# Début de la génération du script SQL
sql_script = """-- Script SQL pour insérer 2 millions d'enregistrements dans public.clients_transactions
BEGIN;\n"""

# Générer les enregistrements
for i in range(1, nombre_enregistrements + 1):
    code_es = random.choice(codes_es)  # Sélectionner un code ES au hasard parmi les codes disponibles
    code_oper = random.choice(codes_oper)  # Sélectionner un code_oper au hasard parmi les codes disponibles
    sql_script += f"INSERT INTO public.clients_transactions (id, id_client, num_phone, first_name, last_name, code_es, dt_naissance, mnt, code_oper) VALUES ({i}, 'ID_{i}', '06000000{i % 1000000:06d}', 'FirstName_{i}', 'LastName_{i}', '{code_es}', '1980-01-01', {100.00 + i % 100}, '{code_oper}');\n"

# Fin du script
sql_script += "COMMIT;\n"

# Sauvegarder le script directement sur Google Drive
with open(chemin_google_drive, 'w') as file:
    file.write(sql_script)

# Afficher un message pour indiquer que le fichier a été enregistré
print(f"Le script SQL a été enregistré sur Google Drive : {chemin_google_drive}")
```
